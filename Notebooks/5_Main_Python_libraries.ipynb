### Arrays


The basic data structure that is the **array**,that is exactly like multi-dimensional arrays (or matrices) in any other language. It consists of one or more dimensions of numbers or chars. 

Unlike lists, the elements of the **array** all have the **same type** (which can be Boolean, integer, real, or complex numbers).

Arrays are made using a function call, and the values are passed in **as** a list, or set of lists for higher dimensions. Arrays can have **as many** dimensions as you like up to a language limit of **40** dimensions. A couple of examples:

```python
myarray = np.array([4, 3, 2])

mybigarray = np.array([[3, 2, 4], [3, 3, 2], [4, 5, 2]])
```

Here are an one-dimensional and a two-dimensional array (which is effectively an array of arrays), respectively. 


An array can be **visualized** by using the `print` command:

```python
print(myarray)
```

```python
print(*myarray)
```

```python
print(mybigarray)
```

Making arrays by writing **directly** the values in the function call is fine **only** for small arrays where the numbers aren't regular, but there are several cases where this is not true. 

There are nice ways to make a set of the more interesting arrays, such as those shown in next slides.

There is one more very **useful** command on arrays, which is the `np.where()` command, which returns the indices where a logical expression is **true**.

As some examples:

```python
x = np.where(a>2)
print(x)
```

returns in the variable `x` the indices where the logical expression is **true**, while:

```python
x = np.where(a>2, 0, 1)
print(x)
```

returns a matrix of the **same size** as `a` that contains 0 in those places where the expression was **true** in a and 1 everywhere **else**.

To chain these conditions together you have to use the **bitwise** logical operations, so that:

```python
indices = np.where((a[:,0]>3) | (a[:,1]<3))
print(indices)
```

**Slice operator** to index a single**row** or **column**: When we use a slice operator and **only** index a single row or column, **NumPy** seems to **turn it** into a list, so that it *stops being either a row or a column*. 

This means that the transpose operator **doesn't** do anything to it, and also means that some of the other behavior can be a little **odd**.

There are a **few** ways around the problem, of which the **two** simplest are: 
1.  listing a start and end for the **slice** even for a single row or column;
2. **explicitly** reshaping it afterward.

```python
c = a[0:1,:]
np.shape(c)
```

```python
c = a[0,:].reshape(1,len(a))
np.shape(c)
```

### Optimization


**SciPy** `optimize` provides functions for *minimizing* (or *maximizing*) objective functions, possibly subject to **constraints**. 

It includes solvers for **nonlinear** problems (with support for both local and global optimization algorithms), linear programing, constrained and nonlinear least-squares, root finding and curve fitting.

**SciPy** sub-packages need to be imported **separately**, for example:

```python
from scipy import optimize as opt
```

However, **SciPy** needs that the **NumPy** library has **already** been loaded.


The `optimize` module includes an interface function that can call the others:

where:
* `fun`: the objective function to be **minimized**.
	
* `x0`: **initial** guess.
	
* `args`: extra arguments passed to the objective function and its derivatives.
	
* `jac`: method for computing the gradient vector (`'2-point'`, `'3-point'`, `'cs'`, `bool` or a callable object).
	
* `hess`: method for computing the Hessian matrix (`'2-point'`, `'3-point'`, `'cs'`, `HessianUpdateStrategy` or a callable object).
	
* `hessp`: Hessian of objective function times an arbitrary vector $p$. 

* `method`: type of solver, a string chosen between:



|  **Solver**        |  **Meaning**                                            |
|:-------------------|:--------------------------------------------------------|
| **'Nelder-Mead'**  | Minimization by using the Nelder-Mead algorithm.        |
| **'Powel'**        | Minimization by using the modified Powell algorithm.    |
| **'CG'**           | Minimization by using the conjugate gradient algorithm. |
| **'BFGS'**         | Minimization by using the BFGS algorithm.               |
| **'Newton-CG'**    | Minimization by using the Newton-CG algorithm.          |
| **'L-BFGS-B'**     | Minimization by using the L-BFGS-B algorithm.           |
| **'TNC'**          | Minimization by using a truncated Newton (TNC) algorithm. |
| **'COBYLA'**       | Minimization by using the Constrained Optimization BY Linear Approximation (COBYLA) algorithm. |
| **'SLSQP'**        | Minimization by using Sequential Least SQuares Programming (SLSQP).  |
| **'trust-constr'** | Minimize a scalar function subject to constraints.      |
| **'dogleg'**       | Minimization by using the dog-leg trust-region algorithm. |
| **'trust-ncg'**    | Minimization by using the Newton conjugate gradient trust-region algorithm. |
| **'trust-exact'**  | Minimization by using a nearly exact trust-region algorithm.  |
| **'trust-krylov'** | Minimization by using a nearly exact trust-region algorithm that only requires matrix vector products with the hessian matrix. |
| **custom**         | A callable object.                                   |


* `bounds`: **bounds** on variables for L-BFGS-B, TNC, SLSQP and trust-constr methods.
	
* `constraints`: constraints definition.
	
* `tol`: *tolerance* for termination.
	
* `callback`: called after each iteration.
	
* `options`: a dictionary of solver **options**.

For using its default value, an argument can be **omitted**.

The optimization **result** is represented by a `OptimizeResult` object. Important attributes are: 
* `x`: the **solution** array;
	
* `success`: a Boolean **flag** indicating if the optimizer exited successfully;
	
* `message`: which describes the cause of the termination.


As an **example** of *unconstrained* optimization, we find the minimum of the function:
$$
f(x) = - e^{-\left( x - 0.7 \right)^2},
$$
starting from the initial guess $x_0 = 0$. We have simply:


```python
def f(x):
    return -np.exp(-(x - 0.7)**2)

x0 = 0.0
opt.minimize(f, x0)
```

As an **example** of *constrained* optimization, we find the minimum of the function:
$$
\begin{split}
f\left( x_1, x_2 \right) = & \sqrt{\left( x_1 - 3 \right)^2 + \left( x_2 - 2 \right)^2}, \\
 s.t. ~ & \left| x_1 \right| + \left| x_2 \right| \leq 1.5,
\end{split}
$$
starting from the initial guess $x_0 = \left[  0, \, 0 \right]$. We have simply:


```python
def f(x):
    return np.sqrt((x[0] - 3)**2 + (x[1] - 2)**2)

def cnst(x):
    return np.atleast_1d(1.5 - np.sum(np.abs(x)))

x0 = np.array([0, 0])
opt.minimize(f, x0, constraints={"fun": cnst, "type": "ineq"})
```

| `rfftshift`, `irfftshift` | Shift the frequency bins of `rfft` so that the zero-frequency is in the middle. |
| `fftfreq`                 | Calculate the frequencies corresponding to the FFT bins in the `fft`.           |


As an example, we evaluate the **FFT** of a signal composed of a sum of two sinusoids and a Gaussian noise:

```python
import numpy as np
from scipy import fft

def signal_samples(t, f1, f2):  # For signal generation
    return (2 * np.sin(2 * np.pi * f1 * t) + 3 * np.sin(2 * np.pi * f2 * t) + 2 * np.random.randn(*np.shape(t)))

Fs = 60; T = 4    # Set Fs and time interval
N = Fs*T          # Number of samples
f1 = 5; f2 = 22   # Choose f1 and f2

t = np.linspace(0, T, N)  # Generate sampled t-axes
x = signal_samples(t, f1, f2)  # Generate the signal

X = fft.fft(x)    # Evaluate the FFT

f = fft.fftfreq(N, 1.0/Fs)  # Generate the f-axis
mask = np.where(f >= 0)     # Mask for positive frequencies


fig, ax = plt.subplots(2, 1, figsize=(12, 10))
# Plot of the time-domain signal
ax[0].plot(t, x, label="Time")
ax[0].grid()
ax[0].set_title("Time domain signal")
ax[0].set_xlabel("$t$", fontsize=14)
ax[0].set_ylabel("$x(t)$", fontsize=14)
ax[0].legend()

# Plot of the frequency-domain Amplitude spectrum
ax[1].plot(f[mask], abs(X[mask])/N, label="Amplitude")
ax[1].grid()
ax[1].set_title("Frequency domain signal")
ax[1].set_xlabel("$f$", fontsize=14)
ax[1].set_ylabel("$|X(f)|/N$", fontsize=14)
ax[1].legend()
```

### Signal Processing: filtering


The output of a **LTI** system can be evaluated by the **convolution**:
$$
y[n] = \sum_{k=0}^N x[k] h[n-k],
$$
or by the **finite difference** equation:
$$
y[n] = \sum_{k=0}^N b_k x[n-k] - \sum_{k=1}^M a_k y[n-k].
$$
These two relationships can be implemented in **SciPy** by the following two commands:

There exist many other methods similar to those of the **MATLAB** Signal Processing Toolbox.

As a simple example, we generate a noisy signal, composed of the sum of two sinusoids and a Gaussian noise:

```python
import numpy as np
from scipy import signal

# Function for signal generation
def signal_samples(t, f1, f2):
    return (2 * np.sin(2 * np.pi * f1 * t) + 3 * np.sin(2 * np.pi * f2 * t) + 0.5 * np.random.randn(*np.shape(t)))

# Settings
Fs = 60; T = 1
N = Fs*T
f1 = 5; f2 = 8

# Generate the noisy signal
t = np.linspace(0, T, N)
x = signal_samples(t, f1, f2)
```

We then **smooth** such a signal with a simple *low-pass* filter (with impulse response equal to $h = [1, 1, 1, ...]/M$):

```python
# Create the impulse response of a LP filter
M = 4
h = np.ones((M,))/M
y = signal.lfilter(h, 1, x)  # Filter the signal

# Plotting the original and smoothed signals
plt.figure()
plt.plot(t, x, label='Original')
plt.plot(t, y, color='red', label='Smoothed')
plt.grid()
plt.title('Smoothing of a signal')
plt.xlabel('t')
plt.ylabel('Amplitude')
plt.legend()
```
